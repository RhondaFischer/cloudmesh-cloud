```python
def parse(table, fgrep=None):
    data = []
    lines = table.strip().split("\n")
    headers = lines[1].strip().split("|")[1:-1]
    header = list(map(str.strip, headers))

    for line in lines[2:]:
        if fgrep is not None and fgrep not in line:
            continue
        line = line.strip()
        if not line.startswith("+"):
            entries = line.split("|")[1:-1]
            entries = list(map(str.strip, entries))
            data.append(entries)

    return header, data
```

```python
def find(data, key):
    header, data = parse(data, fgrep=key)
    df = DataFrame(data, columns=header)
    df['start'] = pd.to_datetime(df['start'])
    df['time'] = df['time'].astype(float)
    return df
```

```python
def plot(df):
    plt.plot(df['start'], df['time'], "*")
    plt.xticks(rotation='vertical')
    plt.ylabel('Seconds')
    plt.xlabel('Date')
```

```python
def histogram(df):
    x = df["time"]
    bins = 20
    n, bins, patches = plt.hist(x, bins, density=1, facecolor='blue', alpha=0.5)
    plt.ylabel('Percentage')
    plt.xlabel('Seconds')
```
